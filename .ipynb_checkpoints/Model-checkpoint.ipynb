## Loading Data

```python

def load_train():
    global df

    curve_dir = train_dir/"curve"
    no_curve_dir = train_dir/"no_curve"
    
    curve_cases = curve_dir.glob("*.png")
    no_curve_cases = no_curve_dir.glob("*.png")
    
    train_data = []
    train_label = []

    for img in curve_cases:
        train_data.append(img)
        train_label.append(1)
    
    for img in no_curve_cases:
        train_data.append(img)
        train_label.append(0)
        
    df = pd.DataFrame(train_data)
    df.columns = ["images"]
    df["labels"] = train_label
    
  #  df = df.sample(frac=1).reset_index(drop=True)
    return df
```

## Generating Data in Batches

```python
def data_gen(data, batch_size):
    # total number of samples in the data
    n = len(data)
    steps = n//batch_size
    
    #numpy array for containing batch data and labels
    batch_data = np.zeros((batch_size, 256, 256, 3), dtype = np.float32)
    batch_label = np.zeros((batch_size,2), dtype = np.float32)
    
    #np arrau for all indices of input data
    indices = np.arange(n)

    i=0
    while True:
        np.random.shuffle(indices)
        count = 0
        next_batch = indices[(i*batch_size) :(i+1)*batch_size]
        
        for j, idx in enumerate(next_batch):
            img_name = data.iloc[idx]["images"]
            label = data.iloc[idx]["images"]
            
            if(label == 0):
                label = 0
            else:
                label = 1
            encoded_label = to_categorical(label, num_classes=2)
            
            img = cv2.imread(str(img_name))
            
            orig_img = img.astype(np.float32)/255
            
            
            batch_data[count] = orig_img
            batch_label[count] = encoded_label
            
            count += 1
            
            if count == batch_size-1:
                break;
        i += 1
        yield batch_data, batch_label
        
        if i>=steps:
            i=0
            
            
```

## CNN Model

```python
model = Sequential()
model.add(Conv2D(32,(3,3), input_shape=(224,224,3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(32, (3,3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
          
model.add(Conv2D(64, (3,3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
          
model.add(Flatten())
          
model.add(Dense(64))
model.add(Activation('relu'))
model.add(Dense(2))
model.add(Activation('softmax'))
```

## Training the Model

```python
batch_size = 10
nb_epochs = 11

#training data generator
train_data_gen = data_gen(data=train_data, batch_size=batch_size)

# defining the number of training steps
nb_train_steps = train_data.shape[0]//batch_size

print("Number of Training Steps  : ",nb_train_steps)
# print("Number of Validation Step : ",len(val_data))
```
